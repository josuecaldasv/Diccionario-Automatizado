```python
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
def automatized_dict(input_path, output_path):
    df = pd.read_excel(input_path)

    # Crea una nueva tabla con la estructura deseada
    new_df = pd.DataFrame(columns=['Variable', 'Variable label', 'Value', 'Value label', 'values_and_labels'])

    # Variables para almacenar los valores y etiquetas
    variable = ''
    variable_label = ''
    values = []
    value_labels = []

    # Itera sobre las filas del DataFrame original
    for index, row in df.iterrows():
        if pd.notnull(row['columna2']):
            if isinstance(row['columna2'], str) and row['columna2'].startswith('Variable ='):
                # Añade la fila anterior solo si los valores no están vacíos
                if variable != '':
                    if values:  # Si existen valores, crea la cadena de valores y etiquetas
                        values_and_labels = '; '.join([f'{v} = {l}' for v, l in zip(values, value_labels)])
                    else:  # Si no hay valores, asigna un espacio en blanco
                        values_and_labels = ' '

                    new_df = new_df.append({
                        'Variable': variable,
                        'Variable label': variable_label,
                        'Value': ';'.join(values),
                        'Value label': ';'.join(value_labels),
                        'values_and_labels': values_and_labels
                    }, ignore_index=True)

                # Reinicia las variables para la nueva variable
                variable = row['columna2'].split('=')[1].strip()
                variable_label = row['columna3'].split('=')[1].strip()
                values = []
                value_labels = []
            elif isinstance(row['columna2'], str) and row['columna2'].startswith('Value ='):
                values.append(row['columna2'].split('=')[1].strip())
                value_labels.append(row['columna3'].split('=')[1].strip())

    # Añade la última fila si los valores no están vacíos
    if variable != '':
        if values:  # Si existen valores, crea la cadena de valores y etiquetas
            values_and_labels = '; '.join([f'{v} = {l}' for v, l in zip(values, value_labels)])
        else:  # Si no hay valores, asigna un espacio en blanco
            values_and_labels = ' '

        new_df = new_df.append({
            'Variable': variable,
            'Variable label': variable_label,
            'Value': ';'.join(values),
            'Value label': ';'.join(value_labels),
            'values_and_labels': values_and_labels
        }, ignore_index=True)

    # Asigna espacios en blanco a las variables sin valores ni etiquetas
    new_df['Value'].fillna('', inplace=True)
    new_df['Value label'].fillna('', inplace=True)
    new_df['values_and_labels'].fillna('', inplace=True)

    # Elimina las filas duplicadas en base a la columna 'Variable'
    new_df = new_df.drop_duplicates(subset='Variable')

    # Exporta la nueva tabla
    new_df.to_excel(output_path, index=False)

    return new_df
```

```python
input_path = r'raw/NdM3_Community_Level_Raw.xlsx'
output_path = r'output/NdM3_Community_Level_Output.xlsx'

automatized_dict( input_path, output_path )
```

```python
# ANTIGUA FUNCIÓN QUE NO INCLUYE VARIABLES SIN VALUES

def automatized_dict( input_path, output_path ):

    df = pd.read_excel( input_path )

    # Crea una nueva tabla con la estructura deseada
    new_df = pd.DataFrame(columns=['Variable', 'Variable label', 'Value', 'Value label', 'values_and_labels'])

    # Itera sobre las filas del DataFrame original
    for index, row in df.iterrows():
        if pd.notnull(row['columna2']):
            if isinstance(row['columna2'], str) and row['columna2'].startswith('Variable ='):
                variable = row['columna2'].split('=')[1].strip()
                variable_label = row['columna3'].split('=')[1].strip()
                value = ''
                value_label = ''
            elif isinstance(row['columna2'], str) and row['columna2'].startswith('Value ='):
                value += row['columna2'].split('=')[1].strip() + ';'
                value_label += row['columna3'].split('=')[1].strip() + ';'
            else:
                continue

        # Añade una nueva fila a la tabla si es una nueva variable
        if index == len(df) - 1 or (pd.isnull(df.iloc[index + 1]['columna2']) or df.iloc[index + 1]['columna2'].startswith('Variable =')):
            # Elimina el último ';' de los valores
            value = value.rstrip(';')
            value_label = value_label.rstrip(';')

            # Añade una nueva fila a la tabla solo si los valores no están vacíos
            if value != '' and value_label != '':
                values_and_labels = '; '.join([f'{v} = {l}' for v, l in zip(value.split(';'), value_label.split(';'))])
                new_df = new_df.append({
                    'Variable': variable,
                    'Variable label': variable_label,
                    'Value': value,
                    'Value label': value_label,
                    'values_and_labels': values_and_labels
                }, ignore_index=True)

    # drip duplicated rows
    new_df = new_df.drop_duplicates( subset = 'Variable' )

    # Exporta la nueva tabla
    new_df.to_excel( output_path, index=False )
    
    return new_df
```
